#### Plot NA value proportions

```python
num_nulls = train_ds.isna().sum()

nulls_pct = pd.DataFrame({'NA_pct': [(x / 1460) * 100 for x in num_nulls], 'colnames': train_ds.columns.values})
nulls_pct.sort_values(by='NA_pct', inplace=True)

ax, fig = plt.subplots(figsize=(10,17))

sns.barplot(data=nulls_pct, x='NA_pct', y='colnames', color='red')
plt.title("Plot of the percentage of NA values in each column")
plt.xlabel("NA Percentage")
plt.ylabel("Feature")
```
